Ground Rules:

1.Combinations must start with an open(L) done

2.Combinations must end with a closed(R) done

3.The number of L and R must be equal, and each must be equal to n. done

4.Each L bracket must have a corresponding R bracket. (See checkvalidity and children) done

5. When adding a R, the number of existing R must not be equal to or more than n, and must have a corresponding L bracket before it.

i.e. L > R

6. When adding an L, the number of existing L must not be equal to or more than n, and R should not be equal to or more than n.

i.e. L < n and R < n
